```python
# Task 1 – Worksheet 2
# Purpose: Given census data, find the State that contains the least number of counties.
# Output: Name of that state as a string.


def answer1():
    
    # Read census data from csv and filter out state info; only want county info.
    census_data = pd.read_csv('census.csv')
    county_data = census_data.loc[census_data['COUNTY'] != 0]
    
    # Group on state name and count number of counties.
    counties_per_state = county_data.groupby('STNAME')['CTYNAME'].count()
    
    # Return the first state that has the minimum number of counties.
    return [x for x in counties_per_state.index if counties_per_state[x] == min(counties_per_state)][0]

answer1()
```

```python
# Task 2 – Worksheet 2
# Purpose: Given census data, find the State that has the highest birth rate over 2010-2015.
# Output: Name of that State as a string, and average births per year.


def answer2():
    
    # Read census data from csv and filter out county info; only want state info.
    census_data = pd.read_csv('census.csv')
    state_data = census_data.loc[census_data['COUNTY'] == 0]
    
    # Sum over the births columns from 2010 to 2015, (not including 2015) divide by 5 years to get average. 
    avg_births = sum([state_data.groupby('STNAME')['BIRTHS' + str(x)].sum() for x in range(2010,2015)]) / 5
    
    # Return the first state that has the highest average birth rate, with that birth rate.
    return [(x, avg_births[x]) for x in avg_births.keys() if avg_births[x] == max(avg_births)][0]

answer2()
```

```python
# Task 3 – Worksheet 2
# Purpose: Given census data, find top 3 most populous counties in each region in 2010.
# Output: DataFrame with region number, state name, county name, and population in 2010; sorted by population descending.

def answer3():
    
    # Read census data from csv and filter out state info; only want county info.
    census_data = pd.read_csv('census.csv')
    county_data = census_data.loc[census_data['COUNTY'] != 0]
    
    top_pop_by_region = []
    
    # looping through each region
    for region in [1,2,3,4]:
        
        # filter for only the counties in that region, sort values descending, and pick out top 3, 
        #    with the necessary info: region (1st column), state (5th), county (6th), population (7th).
        top_pop_by_region.append(county_data.loc[county_data['REGION'] == region].sort_values(by = 'CENSUS2010POP', ascending = False).iloc[0:3,[1,5,6,7]])
        
    # concatenate top 3 from each of four regions, sort again by population descending. Return the resulting DataFrame
    return pd.concat(top_pop_by_region).sort_values(by = 'CENSUS2010POP', ascending = False)
        
answer3()
```

```python
# Task 4 – Worksheet 2
# Purpose: Given census data, find the average birth rate and death rate over 2010 to 2015 for each state.
# Output: DataFrame with state name, average birth rate and death rate.

# Set precision to 3 so table doesn't have too many trailing decimals
pd.set_option('precision', 3)

def answer4():
    
    # Read census data from csv and filter out county info; only want state info.
    census_data = pd.read_csv('census.csv')
    state_data = census_data.loc[census_data['COUNTY'] == 0]
    
    # sum births and deaths over 2010 to 2015 (inclusive). Divide by 6 to get average.
    avg_births = (sum([state_data['BIRTHS' + str(x)] for x in range(2010, 2016)]) / 6).rename('Average Birth Rate')
    avg_deaths = (sum([state_data['DEATHS' + str(x)] for x in range(2010, 2016)]) / 6).rename('Average Death Rate')
    
    # Concatenate average births and deaths horizontally, use the state names as the index. Return resulting DataFrame
    return pd.concat([avg_births, avg_deaths], axis = 1).set_index(pd.Index(state_data['STNAME']))

answer4()
        
    
```

```python
# Task 5 – Worksheet 2
# Purpose: Given census data, find the proportional increase in population due to births (births / total estimated 
#    popultation) and proportional decrease in population due to deaths (deaths / total estimated popultation) each
#    year 2010 to 2015 for each county.
# Output: DataFrame with birth ratio and death ratio for each year, sorted by state name then county name.

# Set precision to 5 since ratios are small.
pd.set_option('precision', 5)

def answer5():
    
    # Read census data from csv and filter out state info; only want county info.
    census_data = pd.read_csv('census.csv')
    county_data = census_data.loc[census_data['COUNTY'] != 0]
    
    # divide birth rate by population estimate for each year 2010 to 2015, same for death ratios.
    birth_ratios = [(county_data['BIRTHS' + str(x)] / county_data['POPESTIMATE' + str(x)]).rename('Birth Ratio ' + str(x)) for x in range(2010, 2016)]
    death_ratios = [(county_data['DEATHS' + str(x)] / county_data['POPESTIMATE' + str(x)]).rename('Death Ratio ' + str(x)) for x in range(2010, 2016)]
    
    # concatenate birth ratios and death ratios horizontally, with hierarchical index (MultiIndex) of state name and 
    #    county name.
    return pd.concat(birth_ratios + death_ratios, axis = 1).set_index(pd.MultiIndex.from_frame(county_data[['STNAME','CTYNAME']]))
    
answer5()
```
